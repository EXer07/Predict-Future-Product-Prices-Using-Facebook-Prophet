# TASK 4: PREPARE THE DATA BEFORE APPLYING FACEBOOK PROPHET TOOL 

```python
avocado_data
```

```python
#เราจะใช้แค่ 2 cloumn คือ AvgagePrice กับ Date
avocado_prophet_df = avocado_data[['Date','AveragePrice']]
```

```python
avocado_prophet_df
```

```python
#เปลี่ยนชื่อ column โดยเปลี่ยนจาก Date เป็น ds และ Average เป็น y
avocado_prophet_df =avocado_prophet_df.rename(columns ={'Date':'ds','AveragePrice':'y'})
```

```python
avocado_prophet_df
```

# TASK 5: UNDERSTAND INTUITION BEHIND FACEBOOK PROPHET

# TASK 6: DEVELOP MODEL AND MAKE PREDICTIONS - PART A

```python
# m มาจาก model
m = Prophet()
m.fit(avocado_prophet_df)
```

```python
# Forcasting into the future ตรง period ใส่ระยะเวลาทำนาย
future = m.make_future_dataframe(periods = 365 )
forecast = m.predict(future)
```

```python
forecast
```

```python
figure = m.plot(forecast , xlabel = 'Date', ylabel = 'Price')
```

```python
figure2 = m.plot_components(forecast)
```
